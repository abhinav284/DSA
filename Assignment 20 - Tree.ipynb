Output2 : 7

 -2

Also, entire tree sum is also 7.

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def findMaxSubtreeSum(node):
    global maxSum
    if node is None:
        return 0
    leftSum = findMaxSubtreeSum(node.left)
    rightSum = findMaxSubtreeSum(node.right)
    currentSum = node.val + leftSum + rightSum
    maxSum = max(maxSum, currentSum)
    return currentSum

def buildBinaryTree(nodes, index):
    if index < len(nodes):
        value = nodes[index]
        if value is None:
            return None

        node = TreeNode(value)
        node.left = buildBinaryTree(nodes, 2 * index + 1)
        node.right = buildBinaryTree(nodes, 2 * index + 2)
        return node
    return None

print("Enter the elements :")
elements = input().split()
nodes = []
for element in elements:
    if element.lower() == "none":
        nodes.append(None)
    else:
        nodes.append(int(element))
root = buildBinaryTree(nodes, 0)
maxSum = float('-inf')
findMaxSubtreeSum(root)
print("Maximum subtree sum:", maxSum)

```

Question-2

Construct the BST (Binary Search Tree) from its given level order traversal.

Example:

Input: arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}

Output: BST:

            7

         /Â  Â  \

       4 Â  Â  12

     /Â  \ Â  Â  /

    3 Â  6Â  8

   / Â   / Â    \

 1Â    5  Â    10

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def constructBST(levelOrder):
    if not levelOrder:
        return None

    root = TreeNode(levelOrder[0])
    queue = [root]
    i = 1
    while i < len(levelOrder):
        current = queue.pop(0)
        left_val = levelOrder[i]
        if left_val is not None:
            current.left = TreeNode(left_val)
            queue.append(current.left)
        i += 1
        if i < len(levelOrder):
            right_val = levelOrder[i]
            if right_val is not None:
                current.right = TreeNode(right_val)
                queue.append(current.right)
            i += 1
    return root

print("Enter the elements :")
elements = input().split()
levelOrder = []
for element in elements:
    if element.lower() == "none":
        levelOrder.append(None)
    else:
        levelOrder.append(int(element))
root = constructBST(levelOrder)
def inorderTraversal(node):
    if node is None:
        return []

    result = []
    result += inorderTraversal(node.left)
    result.append(node.val)
    result += inorderTraversal(node.right)
    return result

inorder = inorderTraversal(root)
print("BST (inorder traversal):", inorder)
```

Given an array of size n. The problem is to check whether the given array can represent the level order traversal of a Binary Search Tree or not.

Examples:

Input1 : arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}

Output1 : Yes

For the given arr[], the Binary Search Tree is:

            7

         /Â  Â  \

       4 Â  Â  12

Input2 : arr[] = {11, 6, 13, 5, 12, 10}

Output2 : No

The given arr[] does not represent the level order traversal of a BST.

```python
from collections import deque

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def checkLevelOrderBST(arr):
    if not arr:
        return "No"
    root = TreeNode(arr[0])
    queue = deque([root])
    i = 1
    while i < len(arr) and queue:
        current = queue.popleft()
        left = None
        right = None
        for j in range(i, len(arr)):
            if arr[j] < current.val:
                left = arr[j]
                break
        for j in range(i, len(arr)):
            if arr[j] > current.val:
                right = arr[j]
                break
        if left is not None:
            current.left = TreeNode(left)
            queue.append(current.left)
        if right is not None:
            current.right = TreeNode(right)
            queue.append(current.right)
        i += 1
    if i < len(arr) or queue:
        return "No" 
    return "Yes"  
print("Enter the elements of the level order traversal:")
elements = input().split()
arr = [int(element) for element in elements]
result = checkLevelOrderBST(arr)
print("Output:", result)
```
