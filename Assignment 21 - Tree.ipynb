Question-1

You are given a binary tree. The binary tree is represented using the TreeNode class. Each TreeNode has an integer value and left and right children, represented using the TreeNode class itself. Convert this binary tree into a binary search tree.

Input:

        10

       /   \

     2      7

   /   \

 8      4

Output:

        8

      /   \

    4     10

  /   \

2      7




```python
class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

def inorder_traversal(node, values):
    if node is None:
        return
    inorder_traversal(node.left, values)
    values.append(node.val)
    inorder_traversal(node.right, values)

def convert_binary_tree_to_bst(root):
    values = []
    inorder_traversal(root, values)

    values.sort()

    def inorder_traversal_replace(node, values):
        if node is None:
            return
        inorder_traversal_replace(node.left, values)
        node.val = values.pop(0)
        inorder_traversal_replace(node.right, values)

    inorder_traversal_replace(root, values)

    return root


def build_binary_tree():
    val = int(input("Enter the value for the root node: "))
    root = TreeNode(val)
    queue = [root]

    while queue:
        current = queue.pop(0)
        left_val = input(f"Enter the value for the left child of {current.val} (-1 if no left child): ")
        if left_val != "-1":
            left = TreeNode(int(left_val))
            current.left = left
            queue.append(left)

        right_val = input(f"Enter the value for the right child of {current.val} (-1 if no right child): ")
        if right_val != "-1":
            right = TreeNode(int(right_val))
            current.right = right
            queue.append(right)

    return root

root = build_binary_tree()
new_root = convert_binary_tree_to_bst(root)

def inorder_traversal_print(node):
    if node is None:
        return
    inorder_traversal_print(node.left)
    print(node.val, end=" ")
    inorder_traversal_print(node.right)

print("Binary Search Tree (inorder traversal):")
inorder_traversal_print(new_root)
```

Question-2:

Given a Binary Search Tree with all unique values and two keys. Find the distance between two nodes in BST. The given keys always exist in BST.

Example:

Consider the following BST:


Input-1:

n = 9

values = [8, 3, 1, 6, 4, 7, 10, 14,13]

node-1 = 6

node-2 = 14

Output-1:

The distance between the two keys = 4

Input-2:

n = 9

values = [8, 3, 1, 6, 4, 7, 10, 14,13]

node-1 = 3

node-2 = 4

**Output-2:**

The distance between the two keys = 2





```python
class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

def build_bst(values):
    root = None
    for val in values:
        root = insert_node(root, val)
    return root

def insert_node(root, value):
    if root is None:
        return TreeNode(value)

    if value < root.val:
        root.left = insert_node(root.left, value)
    elif value > root.val:
        root.right = insert_node(root.right, value)

    return root

def find_distance(root, node1, node2):
    lca = find_lca(root, node1, node2)
    dist1 = find_distance_from_node(lca, node1)
    dist2 = find_distance_from_node(lca, node2)
    distance = dist1 + dist2
    return distance

def find_lca(root, node1, node2):
    if root.val > node1 and root.val > node2:
        return find_lca(root.left, node1, node2)
    elif root.val < node1 and root.val < node2:
        return find_lca(root.right, node1, node2)
    else:
        return root

def find_distance_from_node(root, node):
    distance = 0
    while root.val != node:
        if node < root.val:
            root = root.left
        else:
            root = root.right
        distance += 1
    return distance

n = int(input("Enter the number of values: "))
values = []
for _ in range(n):
    value = int(input("Enter a value: "))
    values.append(value)
root = build_bst(values)
node1 = int(input("Enter node 1: "))
node2 = int(input("Enter node 2: "))
distance = find_distance(root, node1, node2)
print("The distance between the two keys =", distance)
```

Question-3:

Write a program to convert a binary tree to a doubly linked list.

Input:

        10

       /   \

     5     20

           /   \

        30     35

Output:

5 10 30 20 35


```python
class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None


class DoublyLinkedListNode:
    def __init__(self, value):
        self.val = value
        self.prev = None
        self.next = None


def convert_binary_tree_to_doubly_linked_list(root):
    if root is None:
        return None
    def inorder_traversal(node, prev):
        nonlocal head, tail

        if node is None:
            return prev
        prev = inorder_traversal(node.left, prev)

        if prev is None:
            head = node  
        else:
            prev.right = node
            node.left = prev
        prev = node  
        prev = inorder_traversal(node.right, prev)

        if prev is None:
            tail = node  
        return prev
    head = None
    tail = None
    inorder_traversal(root, None)
    return head

def build_binary_tree():
    val = int(input("Enter the value for the root node: "))
    root = TreeNode(val)
    queue = [root]
    while queue:
        current = queue.pop(0)
        left_val = input(f"Enter the value for the left child of {current.val} (-1 if no left child): ")
        if left_val != "-1":
            left = TreeNode(int(left_val))
            current.left = left
            queue.append(left)

        right_val = input(f"Enter the value for the right child of {current.val} (-1 if no right child): ")
        if right_val != "-1":
            right = TreeNode(int(right_val))
            current.right = right
            queue.append(right)

    return root
root = build_binary_tree()
head = convert_binary_tree_to_doubly_linked_list(root)
current = head
while current is not None:
    print(current.val, end=" ")
    current = current.right
```

Question-4:

Write a program to connect nodes at the same level.

Input:

        1

      /   \

    2      3

  /   \   /   \

4     5 6    7

Output:

1 → -1

2 → 3

3 → -1

4 → 5

5 → 6

6 → 7

7 → -1

```python
class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None
        self.next = None


def connect_nodes_at_same_level(root):
    if root is None:
        return
    queue = [root]
    while queue:
        level_size = len(queue)
        for i in range(level_size):
            node = queue.pop(0)
            if i < level_size - 1:
                node.next = queue[0]
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)

def print_connections(root):
    current = root
    while current:
        print(current.val, end=" → ")
        if current.next:
            current = current.next
        else:
            print("-1")
            current = current.left

def build_binary_tree():
    val = int(input("Enter the value for the root node: "))
    root = TreeNode(val)
    queue = [root]
    while queue:
        current = queue.pop(0)
        left_val = input(f"Enter the value for the left child of {current.val} (-1 if no left child): ")
        if left_val != "-1":
            left = TreeNode(int(left_val))
            current.left = left
            queue.append(left)
        right_val = input(f"Enter the value for the right child of {current.val} (-1 if no right child): ")
        if right_val != "-1":
            right = TreeNode(int(right_val))
            current.right = right
            queue.append(right)
    return root

root = build_binary_tree()
connect_nodes_at_same_level(root)
print_connections(root)
```
