1. Write a Python program to reverse a string without using any built-in string reversal functions.

```python
def reverse_str(string):
    reversed_str = ""
    for i in range(len(string) - 1, -1, -1):
        reversed_str += string[i]
    return reversed_str

input_str = input("string: ")
reversed_str = reverse_str(input_str)
print("Reversed str:", reversed_str)
```

3. Write a program to find the largest element in a given list.


```python
def find_largest_element(lst):
    if not lst:
        return None

    largest = lst[0]

    for element in lst:
        if element > largest:
            largest = element

    return largest

input_list = input("Enter in arr: ").split()
input_list = [int(num) for num in input_list]  
largest_element = find_largest_element(input_list)
print("The largest element is:", largest_element)
```

5. Write a Python program to find the second largest number in a list.


```python
def second_largest_element(lst):
    if len(lst) < 2:
        return None

    largest = max(lst[0], lst[1])
    second_largest = min(lst[0], lst[1])

    for num in lst[2:]:
        if num > largest:
            second_largest = largest
            largest = num
        elif num > second_largest and num != largest:
            second_largest = num

    return second_largest

input_list = input("Enter a input: ").split()
input_list = [int(num) for num in input_list]  
second_largest = second_largest_element(input_list)

if second_largest is not None:
    print("second largest element:", second_largest)
else:
    print("No element")
```

13. Write a Python program to generate all permutations of a given string.



```python
def generate_permutations(string):
    if len(string) == 0:
        return []

    if len(string) == 1:
        return [string]
    permutations = []
    for i in range(len(string)):
        current_char = string[i]
        remaining_chars = string[:i] + string[i+1:]
        sub_permutations = generate_permutations(remaining_chars)
        for sub_permutation in sub_permutations:
            permutations.append(current_char + sub_permutation)
    return permutations

input_string = input("Enter a string: ")
permutations = generate_permutations(input_string)
print("permutations of the string:")
for permutation in permutations:
    print(permutation)
```

14. Implement a function to calculate the Fibonacci sequence up to a given number of terms.


```python
def fibonacci_sequence(n):
    sequence = []
    if n >= 1:
        sequence.append(0)
    if n >= 2:
        sequence.append(1)
    for i in range(2, n):
        next_number = sequence[i - 1] + sequence[i - 2]
        sequence.append(next_number)
    return sequence

num_terms = int(input("Enter the number: "))
fib_sequence = fibonacci_sequence(num_terms)
print("Fibonacci sequence up to", num_terms, "terms:")
print(fib_sequence)
```

15. Write a program to find the median of a list of numbers.


```python
def find_median(lst):
    sorted_lst = sorted(lst)
    n = len(sorted_lst)
    if n % 2 == 0:
        middle_right = n // 2
        middle_left = middle_right - 1
        median = (sorted_lst[middle_left] + sorted_lst[middle_right]) / 2
    else:
        middle = n // 2
        median = sorted_lst[middle]
    return median

input_list = input("Enter a numbers: ").split()
input_list = [float(num) for num in input_list]  # Convert input to floats
median = find_median(input_list)
print("The median of the list is:", median)
```

18. Implement a function to find the maximum subarray sum in a given list.


```python
def find_maximum_subarray_sum(lst):
    max_sum = lst[0] 
    current_sum = lst[0] 

    for num in lst[1:]:
        current_sum = max(num, current_sum + num)
        max_sum = max(max_sum, current_sum)

    return max_sum

# Example usage
input_list = input("Enter numbers: ").split()
input_list = [int(num) for num in input_list]  
max_subarray_sum = find_maximum_subarray_sum(input_list)
print("The maximum subarray sum:", max_subarray_sum)
```

```python
def non_repeating_char(string):
    char_counts = {}

    for char in string:
        if char in char_counts:
            char_counts[char] += 1
        else:
            char_counts[char] = 1
    for char in string:
        if char_counts[char] == 1:
            return char
    return None
input_string = input("Enter a string: ")
first_non_repeating_char = non_repeating_char(input_string)
if first_non_repeating_char:
    print("non-repeating character is:", first_non_repeating_char)
else:
    print("There are no non-repeating characters in the string.")
```

23. Write a program to find the prime factors of a given number.



```python
def find_prime_factors(number):
    prime_factors = []
    divisor = 2
    while divisor <= number:
        if number % divisor == 0:
            prime_factors.append(divisor)
            number //= divisor
        else:
            divisor += 1
    return prime_factors
input_number = int(input("Enter a number: "))
prime_factors = find_prime_factors(input_number)
print("Prime factors of", input_number, "are:", prime_factors)
```

25. Write a Python program to merge two sorted lists into a single sorted list.


```python
def merge_sorted_lists(list1, list2):
    merged_list = []
    i = j = 0

    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            merged_list.append(list1[i])
            i += 1
        else:
            merged_list.append(list2[j])
            j += 1
    while i < len(list1):
        merged_list.append(list1[i])
        i += 1
    while j < len(list2):
        merged_list.append(list2[j])
        j += 1

    return merged_list
input_list1 = input("Enter the 1st elements : ").split()
input_list2 = input("Enter the 2nd elements : ").split()
list1 = [int(num) for num in input_list1]
list2 = [int(num) for num in input_list2]
merged_list = merge_sorted_lists(list1, list2)
print("Merged sorted list:", merged_list)
```

26. Implement a function to find the mode of a list of numbers.


```python
from collections import Counter

def find_mode(lst):
    counter = Counter(lst)
    mode_count = max(counter.values())
    mode = [num for num, count in counter.items() if count == mode_count]
    return mode

input_list = input("Enter numbers: ").split()
input_list = [int(num) for num in input_list]  

mode = find_mode(input_list)

if len(mode) == len(input_list):
    print("The list has no mode.")
else:
    print("The mode(s) of the list:", mode)
```

28. Implement a function to calculate the square root of a given number.


```python
def calculate_square_root(number):
    if number < 0:
        raise ValueError("Square root is not for negative numbers.")
    if number == 0:
        return 0
    x = number / 2
    precision = 0.0001
    while abs(x * x - number) > precision:
        x = (x + number / x) / 2
    return x

input_number = float(input("Enter a number: "))
square_root = calculate_square_root(input_number)
print("square root of", input_number, "is:", square_root)
```

30. Implement a function to find the minimum element in a rotated sorted list.


```python
def find_minimum(nums):
    left = 0
    right = len(nums) - 1
    if nums[left] <= nums[right]:
        return nums[left]
    while left < right:
        mid = (left + right) // 2
        if nums[mid] > nums[right]:
            left = mid + 1
        else:
            right = mid
    return nums[left]

input_list = input("Enter numbers : ").split()
input_list = [int(num) for num in input_list] 

minimum = find_minimum(input_list)
print("minimum element list is:", minimum)
```

32. Implement a function to calculate the power of a number using recursion.


```python
def calculate_power(base, exponent):
    if exponent == 0:
        return 1
    elif exponent > 0:
        return base * calculate_power(base, exponent - 1)
    else:
        return 1 / base * calculate_power(base, exponent + 1)
input_base = float(input("Enter the base: "))
input_exponent = int(input("Enter the exponent: "))
result = calculate_power(input_base, input_exponent)
print("The result of", input_base, "raised to the power of", input_exponent, "is:", result)
```

33. Write a Python program to remove duplicates from a list while preserving the order.


```python
def preserve_order(lst):
    unique_list = []
    seen = set()
    for item in lst:
        if item not in seen:
            unique_list.append(item)
            seen.add(item)
    return unique_list

input_list = input("Enter elements: ").split()
unique_list = preserve_order(input_list)
print("duplicates removed whilepreserving order:", unique_list)
```

34. Implement a function to find the longest common prefix among a list of strings.


```python
def longest_common_prefix(strings):
    if not strings:
        return ""
    min_length = min(len(string) for string in strings)
    for i in range(min_length):
        char = strings[0][i] 
        if not all(string[i] == char for string in strings):
            return strings[0][:i]  
    return strings[0][:min_length]  

input_strings = input("Enter a list of strings: ").split()
find_longest_common_prefix = longest_common_prefix(input_strings)
print("Longest common prefix:", find_longest_common_prefix)
```

38. Implement a function to find the missing number in a given list of consecutive numbers.


```python
def find_missing_number(numbers):
    n = len(numbers) + 1
    expected_sum = (n * (numbers[0] + numbers[-1])) // 2
    actual_sum = sum(numbers)
    missing_number = expected_sum - actual_sum
    return missing_number

user_input = input("Enter a list of consecutive numbers: ")
numbers = list(map(int, user_input.split()))
missing_number = find_missing_number(numbers)
print("The missing number is:", missing_number)
```
